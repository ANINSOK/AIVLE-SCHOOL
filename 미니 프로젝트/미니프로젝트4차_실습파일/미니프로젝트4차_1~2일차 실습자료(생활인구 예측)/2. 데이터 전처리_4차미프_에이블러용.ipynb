```python
# 학습 데이터 로딩
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings

# 시각화 한글폰트 설정
plt.rc('font', family='Malgun Gothic')
sns.set(font="Malgun Gothic",#"NanumGothicCoding", 
        rc={"axes.unicode_minus":False}, # 마이너스 부호 깨짐 현상 해결
        style='darkgrid')

# 경고 메시지 숨기기
warnings.filterwarnings("ignore", category=UserWarning)
```

```python
df_total = pd.read_csv('../data/df_total.csv')
df_test = pd.read_csv('../data/df_test.csv')
```

# 1. 데이터 전처리

#### [실습문제1] 열 추가 
* 기존 '기준일ID'열에서 날짜 데이터를 추출하여 새로운 열로 추가 합니다.
    - pd.to_datetime() 함수를 사용해 '기준일ID'열을 datetime 형식으로 변경하겠습니다.
    - year, month, day 열로 기존 데이터 프레임에 연월일 정보를 저장합니다.
    - 데이터 프레임 열에 대해 날짜 정보를 얻을 때에는 **dt.year**, **dt.month**, **dt.day** 속성을 사용합니다. 

```python
# 아래에 실습코드를 작성하고 결과를 확인합니다.
df_total['기준일ID'] = pd.to_datetime(df_total['기준일ID'])
df_test['기준일ID'] = pd.to_datetime(df_test['기준일ID'])
```

```python
# 아래에 실습코드를 작성하고 결과를 확인합니다.
df_total['Year'] = df_total['기준일ID'].dt.year
df_total['Month'] = df_total['기준일ID'].dt.month
df_total['Day'] = df_total['기준일ID'].dt.day

# 아래에 실습코드를 작성하고 결과를 확인합니다.
df_test['Year'] = df_test['기준일ID'].dt.year
df_test['Month'] = df_test['기준일ID'].dt.month
df_test['Day'] = df_test['기준일ID'].dt.day
```

#### [실습문제2] 불필요한 열 제거 및 열 이름 변경
* df_total과 df_test에서 '총생활인구수'외에 불필요한 열을 제거 하도록 하겠습니다.
* 시간 데이터를 남기고, '시간대구분' column 을 'hour'로 rename 하도록 하겠습니다. 

```python
# 아래에 실습코드를 작성하고 결과를 확인합니다.
df_total = df_total[['기준일ID', '시간대구분', '총생활인구수','Year', 'Month', 'Day']]
df_test = df_test[['기준일ID', '시간대구분', '총생활인구수','Year', 'Month', 'Day']]
```

```python
# 아래에 실습코드를 작성하고 결과를 확인합니다.
# '시간대구분' 열의 이름을 'hour'로 변경
df_total.rename(columns={'시간대구분': 'hour'}, inplace=True)
df_test.rename(columns={'시간대구분': 'hour'}, inplace=True)
```

#### [실습문제3] 시프트 & 롤링
* shift()를 활용하여 이동할 행 수를 지정합니다.
* rolling()을 활용하여 추가할 이동평균 수를 지정합니다.

```python
# '총생활인구수' 열을 1행씩 아래로 이동
# df_total['yesterday_총생활인구수'] = df_total['총생활인구수'].shift(1)
# print(df_total[['총생활인구수', 'yesterday_총생활인구수']].head())
```

```python
# '총생활인구수' 열을 1행씩 아래로 이동
df_total['내일_총생활인구수'] = df_total['총생활인구수'].shift(-24)
df_test['내일_총생활인구수'] = df_test['총생활인구수'].shift(-24)
```

```python
# '총생활인구수' 열을 6시간 간격으로 이동 평균 계산
rolling_mean = df_total['총생활인구수'].rolling(window=3, min_periods=1).mean()
df_total['rolling_mean_6hour_총생활인구수'] = rolling_mean
print(df_total[['총생활인구수', 'rolling_mean_6hour_총생활인구수']].head())

# '총생활인구수' 열을 3시간 간격으로 이동 평균 계산
rolling_mean = df_test['총생활인구수'].rolling(window=3, min_periods=1).mean()
df_test['rolling_mean_6hour_총생활인구수'] = rolling_mean
print(df_test[['총생활인구수', 'rolling_mean_6hour_총생활인구수']].head())
```

```python
# 계절 변수 생성
def get_season(month):
    if 3 <= month <= 5:
        return '봄'
    elif 6 <= month <= 8:
        return '여름'
    elif 9 <= month <= 11:
        return '가을'
    else:
        return '겨울'
df_total['한달전_총생활인구수'] = df_total['총생활인구수'].shift(24 * 30)
df_total['계절'] = df_total['Month'].apply(get_season)
df_total['요일'] = df_total['기준일ID'].dt.dayofweek
df_total['주말'] = df_total['요일'].apply(lambda x: 1 if x >= 5 else 0)
df_total['시간대'] = df_total['hour'].apply(lambda x: '아침' if 5 <= x < 12 else '오후' if 12 <= x < 18 else '저녁')

df_test['한달전_총생활인구수'] = df_test['총생활인구수'].shift(24 * 30)
df_test['계절'] = df_test['Month'].apply(get_season)
df_test['요일'] = df_test['기준일ID'].dt.dayofweek
df_test['주말'] = df_test['요일'].apply(lambda x: 1 if x >= 5 else 0)
df_test['시간대'] = df_test['hour'].apply(lambda x: '아침' if 5 <= x < 12 else '오후' if 12 <= x < 18 else '저녁')
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
df_total['계절'] = label_encoder.fit_transform(df_total['계절'])
df_total['시간대'] = label_encoder.fit_transform(df_total['시간대'])

df_test['계절'] = label_encoder.fit_transform(df_test['계절'])
df_test['시간대'] = label_encoder.fit_transform(df_test['시간대'])
```

#### [실습문제4] 결측치 처리 및 불필요한 열 제거
* 롤링 & 시프트 과정에서 발생한 결측치를 처리하도록 하겠습니다.
* 불필요한 열을 추가로 제거 하겠습니다.(시간 관련 행들)

```python
# 아래에 실습코드를 작성하고 결과를 확인합니다.
df_total.dropna(inplace=True)
df_total.reset_index(inplace=True, drop=True)
df_test.dropna(inplace=True)
df_test.reset_index(inplace=True, drop=True)
```

```python
df_total.drop(columns='기준일ID', inplace=True)
df_test.drop(columns='기준일ID', inplace=True)
df_total.drop(columns='Year', inplace=True)
df_test.drop(columns='Year', inplace=True)
df_total.drop(columns='Month', inplace=True)
df_test.drop(columns='Month', inplace=True)
df_total.drop(columns='Day', inplace=True)
df_test.drop(columns='Day', inplace=True)
df_total.drop(columns='주말', inplace=True)
df_test.drop(columns='주말', inplace=True)
df_total.drop(columns='hour', inplace=True)
df_test.drop(columns='hour', inplace=True)
```
