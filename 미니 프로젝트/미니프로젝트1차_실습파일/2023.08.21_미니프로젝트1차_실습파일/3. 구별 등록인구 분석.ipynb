#### [3.2.2] 원하는 데이터 형태로 저장하기

* head, tail 등을 살펴본 결과를 토대로 header 값을 적절히 지정하여 데이터를 다시 불러오고 필요한 컬럼만 저장하세요

```python
#read_csv 재실행. 단, header값을 적절히 지정
seoul_people = seoul_people.T
seoul_people.reset_index()
t1 = seoul_people['자치구']
t2 = seoul_people['인구.1']
t3 = seoul_people['인구.2']
t4 = seoul_people['인구']
t5 = seoul_people['65세이상고령자']
#pd.merge(t1, t2, how='left')
tmp1 = pd.concat([t1, t2], axis=1)
tmp2 = pd.concat([tmp1, t3], axis=1)
tmp3 = pd.concat([tmp2, t4], axis=1)
tmp4 = pd.concat([tmp3, t5], axis=1)
```

```python
# 다시 불러온 데이터프레임에서 여러분 각자가 사용하고자 하는 컬럼만 가져오세요.
tmp4.rename(columns={'인구.1':'남자합계'}, inplace=True)
tmp4.rename(columns={'인구.2':'여자합계'}, inplace=True)
tmp4.rename(columns={'인구':'인구합계'}, inplace=True)
tmp4.drop([0,2], axis=0, inplace=True)
```

```python
# 해당 데이터프레임을 csv 파일로 저장하세요.
tmp4 = tmp4.reset_index(drop=True)
tmp4
# tmp4.to_csv('df_seoul_people.csv')
```
